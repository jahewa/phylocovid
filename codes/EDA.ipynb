It worked. Now we have to allign them. But first, save them as *fasta*: 

```python
from Bio.SeqRecord import SeqRecord


seq_records = []
for i in range(len(inf)):
    record = SeqRecord(spike_proteins[i], id=inf[i])
    seq_records.append(record)
seq_records
```

```python
len(seq_records)
```

```python
SeqIO.write(seq_records, "spike_proteins.fasta", "fasta")
```
