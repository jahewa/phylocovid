```python
from Bio import Phylo
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#MAPLE Histogramm
```

```python
tree = Phylo.read('./MAPLE-main/files/MAPLE_output_tree.tree', 'newick')
```

```python
tree
```

```python
depths = tree.depths()
```

```python
depths
```

```python
depths.keys()
```

```python
df = pd.DataFrame.from_dict(depths, orient='index')
df.head()
```

```python
df.info()
```

```python
df.hist(color="#E9884C", figsize=(12,8), bins=25)
plt.title("Brunch lengths")
plt.ylabel("Count")
plt.xlabel("Length")
plt.xticks(np.arange(0, 0.005, 0.0005))
```
